```python
import numpy as np
import matplotlib.pyplot as plt

# font_data.csv 파일을 읽어 데이터 로드
csv_file = "../../Data/font_data.csv"  # CSV 파일 경로
with open(csv_file, "r") as f:
    data_list = f.readlines()

# 데이터 확인 및 변환
all_values = data_list[0].strip().split(',')

# 결측값 처리: '-999'를 float으로 변환
all_values = [float(x) if x != '-999' else -999 for x in all_values]
print(all_values)
# 이미지 데이터 변환 및 시각화
image_array = np.asfarray(all_values[1:]).reshape((28, 28))
plt.imshow(image_array, cmap='gray', interpolation='None')
plt.show()
```

```python
from neuralNetwork import neuralNetwork

# 신경망 초기화
nn = neuralNetwork(
    input_nodes=784,
    input_data=all_values[1:],
    hidden_layers=[512, 256, 160],
    output_nodes=154,
    learning_rate=0.01,
    weight_init="He",
    activation="ReLU",
    output_activation="softmax"
)

# 각 레이어의 정보 출력
def test_layers(nn):
    """신경망의 각 레이어 정보 출력"""
    print("=" * 50)
    print("Neural Network Layers Information:")
    for i, layer in enumerate(nn.layers):
        print(f"Layer {i + 1}:")
        print(f"  Number of Nodes: {layer.layer_size}")
        if layer.weights:
            print(f"  Weight Matrix Size: {len(layer.weights)} x {len(layer.weights[0])}")
            print(f"  Weight example {layer.weights[0][0]}")
        else:
            print("  No weights (likely input layer or uninitialized).")
        print("-" * 50)
    print("=" * 50)

# 테스트 실행
test_layers(nn)
```
