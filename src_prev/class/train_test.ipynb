```python
import numpy as np
import matplotlib.pyplot as plt

# font_data.csv 파일을 읽어 데이터 로드
csv_file = "../../Data/font_data.csv"  # CSV 파일 경로
with open(csv_file, "r") as f:
    data_list = f.readlines()

# 총 데이터 개수
num_data = len(data_list)

# 서브플롯 행과 열 계산 (여기서는 28x55로 배치하여 전체 1540개를 표시)
rows = 28
cols = 55

# 서브플롯 크기 설정
fig, axes = plt.subplots(rows, cols, figsize=(cols, rows))
fig.tight_layout(pad=0.1)

# 각 데이터를 순회하며 시각화
for i, data in enumerate(data_list):
    # 데이터 확인 및 변환
    all_values = data.strip().split(',')
    # 결측값 처리: '-999'를 float으로 변환
    all_values = [float(x) if x != '-999' else -999 for x in all_values]
    
    # 이미지 데이터 변환 및 시각화
    image_array = np.asfarray(all_values[1:]).reshape((28, 28))
    row, col = divmod(i, cols)
    axes[row, col].imshow(image_array, cmap='gray', interpolation='None')
    axes[row, col].axis('off')  # 축 숨기기

plt.show()
```
